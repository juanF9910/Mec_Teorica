Tomamos el sistema coordenado de acuerdo a la imagen anterior. Para este  sistema tenemos los parámetros constantes $l_1$,$l_2$, $m_1$, $m_2$ que corresponden a la longitud de los hilos y a la masa de las partículas respectivamente. 

El número de grados de libertad para el sistema es $$n=3N-k=2$$

$N=2$ es el número de partículas y   $k=4$ es el número de ligaduras . Las ligaduras  todas holónomas son las siguientes:

$$z_1=0$$

$$z_2=0$$

$$x_1^{2}+y_1^{2}=l_1$$

$$x_2^{2}+y_2^{2}=l_2$$

en las anteriores ligaduras se toma el $x_1,y_1$ desde el origen de coordenadas y el $x_2,y_2$ desde la masa $m_1$




**Lagrangiano**

Llamemos a los dos grados de libertad $\phi_1$,$\phi_2$ tal como aparece en la imagen. Establescamos el lagrangiano del sistema:

$$\mathcal{L}=T_1+T_2-(U_1+U_2)=T-U$$




*   Para la partícula 1

$$T_1=\frac{1}{2}m_1(l_1 \dot{\phi}_1)^{2}$$




$$U_1=-m_1gl_1\cos \phi_1$$




*   Para la partícula 2

sus posiciones respecto al origen son 

$$x_2=l_1 \sin \phi_1 + l_2 \sin \phi_2 ~~~~~~~~~~ y_2=l_1 \cos \phi_1 + l_2 \cos \phi_2$$

$$\dot{x}_2=l_1 \dot{\phi}_1 \cos \phi_1 +l_2 \dot{\phi}_2 \cos \phi_2  ~~~~~~~~~~ \dot{y}_2=- \dot{\phi}_1 l_1 \sin \phi_1 - \dot{\phi}_2 l_2 \sin \phi_2$$

$$\dot{x}_2^{2}=l_1^{2} \dot{\phi}_1^{2} \cos^{2} \phi_1 +l_2^{2} \dot{\phi}_2^{2} \cos^{2} \phi_2+2 l_1l_2 \dot{\phi}_1 \dot{\phi}_2 \cos \phi_1 \cos \phi_2  ~~~~~~~~~~ \dot{y}_2^{2}= \dot{\phi}_1^{2} l_1^{2} \sin^{2} \phi_1 +\dot{\phi}_2^{2} l_2^{2} \sin^{2} \phi_2+ 2  l_1 l_2  \dot{\phi}_1  \dot{\phi}_2 \sin \phi_1  \sin \phi_2$$

$$T_2=\frac{1}{2}m_2 \left(\dot{x}_2^{2} + \dot{y}_2^{2} \right)=\frac{1}{2}m_2 \left[ l_1^{2} \dot{\phi}_1^{2}+ l_2^{2} \dot{\phi}_2^{2}+2l_1l_2 \dot{\phi}_1 \dot{\phi}_2 \cos (\phi_1-\phi_2)\right]$$

$$U_2=-m_2g(l_1\cos \phi_1 +l_2 \cos \phi_2)$$


El lagrangiano del sistema sería 

$$\mathcal{L}=\frac{1}{2}m_1(l_1 \dot{\phi}_1)^{2}+\frac{1}{2}m_2 \left[ l_1^{2} \dot{\phi}_1^{2}+ l_2^{2} \dot{\phi}_2^{2}+2l_1l_2 \dot{\phi}_1 \dot{\phi}_2 \cos (\phi_1-\phi_2)\right]+m_1gl_1\cos \phi_1+m_2g(l_1\cos \phi_1 +l_2 \cos l_2)$$


$$=\frac{1}{2}(m_1+m_2)l_1^{2}\dot{\phi}_1^{2}+\frac{1}{2}m_2l_2^{2}\dot{\phi}_2^{2}+m_2l_1l_2 \dot{\phi}_1 \dot{\phi}_2 \cos (\phi_1-\phi_2)+(m_1+m_2)gl_1\cos \phi_1+m_2gl_2\cos \phi_2$$





**Hamiltoniano**

Como este es un sistema conservativo entonces $\mathcal{H}=T+U$

$$p_{\phi_1}=\frac{\partial \mathcal{L}}{\partial \dot{\phi}_1 }=(m_1+m_2)l_1^{2}\dot{\phi}_1+m_2l_1l_2\dot{\phi}_2cos(\phi_1-\phi_2)$$

$$p_{\phi_2}=\frac{\partial \mathcal{L}}{\partial \dot{\phi}_2 }=m_2l_2^{2}\dot{\phi}_2+m_2l_1l_2\dot{\phi}_1cos(\phi_1-\phi_2)$$

Las ecuaciones anteriores las podemos escribir juntas como:


$$\begin{bmatrix} 
p_{\phi_1}\\
p_{\phi_2}
\end{bmatrix}=\begin{bmatrix} 
(m_1+m_2)l_1^{2}& m_2l_1l_2 \cos (\phi_1-\phi_2)\\
m_2l_1l_2 \cos (\phi_1-\phi_2)&m_2l_2^{2}
\end{bmatrix} \begin{bmatrix} 
\dot{\phi}_1\\
\dot{\phi}_2
\end{bmatrix}$$

Para expresar las rapideces generalizadas en términos de los momentos, sólo hallamos la inversa de la matriz de transformación $2\times 2$: Utiliando lo siguiente.

Si $$A=\begin{bmatrix}
a&b\\
c&d
 \end{bmatrix} \Rightarrow A^{-1}=\frac{1}{det(A)}\begin{bmatrix}
d&-b\\
-c&a
 \end{bmatrix} $$

El determinante en este caso es:

$$det=l_1^{2}l_2^{2}m_2(m_1+m_2)-m_2^{2}l_1^{2}l_2^{2}\cos^{2}(\phi_1-\phi_2)$$

Por lo tanto tenemos:

$$ \begin{bmatrix} 
\dot{\phi}_1\\
\dot{\phi}_2
\end{bmatrix} =\begin{bmatrix} 
\frac{1}{l_1^{2}\left[(m_1+m_2)-m_2\cos^{2}(\phi_1-\phi_2) \right]}& -\frac{\cos (\phi_1-\phi_2)}{l_1l_2 [(m_1+m_2)-m_2\cos^{2}(\phi_1-\phi_2)}\\
-\frac{\cos (\phi_1-\phi_2)}{l_1l_2 [(m_1+m_2)-m_2\cos^{2}(\phi_1-\phi_2)}& \frac{(m_1+m_2)}{m_2l_2^{2}[(m_1+m_2)-m_2\cos^{2}(\phi_1-\phi_2)]}
\end{bmatrix} \begin{bmatrix} 
p_{\phi_1}\\
p_{\phi_2}
\end{bmatrix}$$

 Reemplazando en la expresión de la energía cinética
 
 
$$T=\frac{1}{2}\left(p_{\phi_1} \dot{\phi}_1+ p_{\phi_2} \dot{\phi}_2 \right)$$

Tenemos que el hamiltoniano del sistema es:

$$\mathcal{H}= \frac{p_{\phi_1}^{2}}{2l_1^{2}\left[(m_1+m_2)-m_2\cos^{2}(\phi_1-\phi_2) \right]}  -\frac{ p_{\phi_1} p_{\phi_2}\cos (\phi_1-\phi_2)}{l_1l_2 [(m_1+m_2)-m_2\cos^{2}(\phi_1-\phi_2)]}+$$$$\frac{p_{\phi_2}^{2}(m_1+m_2)}{2m_2l_2^{2}[(m_1+m_2)-m_2\cos^{2}(\phi_1-\phi_2)]}-(m_1+m_2)gl_1\cos \phi_1-m_2gl_2\cos \phi_2$$

```python
from sympy import *
pphi1,pphi2,l1,l2,m1,m2,phi1,phi2,g=symbols('p_{\phi_1},p_{\phi_2},l_1,l_2,m_1,m_2,\phi_1,\phi_2,g')
H=(pphi1**2)/(2*l1**2*((m1+m2)-m2*(cos(phi1-phi2))**2))-(pphi1*pphi2*cos(phi1-phi2))/(l1*l2*((m1+m2)-m2*(cos(phi1-phi2))**2))+(pphi2**2*(m1+m2))/(2*m2*l2**2*((m1+m2)-m2*(cos(phi1-phi2))**2))-(m1+m2)*g*l1*cos(phi1)-m2*g*l1*cos(phi2)
```

**Ecuaciones de movimiento de Hamilton**

Determinemos ahora las ecuaciones canónicas de Hamilton a resolver para hallar las $4$ variables dinámicas del problema: $\phi_1,\phi_2,p_{\phi_1},p_{\phi_2}$

$$\dot{\phi}_1=\frac{\partial \mathcal{H}}{\partial p_{\phi_1} }=$$

```python
diff(H,pphi1,1)
```

$$\dot{\phi}_2=\frac{\partial \mathcal{H}}{\partial p_{\phi_2} }=$$

```python
diff(H,pphi2,1)
```

$$\dot{p}_{\phi_1}=-\frac{\partial \mathcal{H}}{\partial \phi_1 }=$$

```python
-1*diff(H,phi1,1)
```

$$\dot{p}_{\phi_2}=-\frac{\partial \mathcal{H}}{\partial \phi_2 }=$$

```python
-1*diff(H,phi2,1)
```

**Solución de las ecuaciones de movimiento**

```python
%matplotlib notebook

import numpy as np
from numpy import *
from math import *
from numpy.linalg import *
import matplotlib.pyplot as plt
from matplotlib import cm
from scipy.integrate import odeint
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.animation as animation

from numpy import *
from scipy.integrate import odeint 
import matplotlib.pyplot as plt
import matplotlib


plt.rcParams['figure.figsize'] = 10, 6 #ancho, alto
font = {'weight' : 'bold',
        'size'   : 15}

matplotlib.rc('font', **font)
```

**Puede cambiar los parámetros y condiciones iniciales del sistema en la siguiente celda**

```python
############################### Parámetros del sistema, en el sistema internacional
m1=0.3 #kg
m2=0.2  #kg
l1=0.20  #m
l2=0.10  #m
g=9.8  #m/s**2

#Condiciones iniciales
[phi01,phi02,pphi01,pphi02]=[1,0.1745,0,0]  #valores iniciales de phi1, phi2 en radianes y  P_phi1, P_phi2 
```

```python
def func(A,t):
    phi1,phi2,pphi1,pphi2=A
    return [-1*(pphi2*cos(phi1-phi2))/(l1*l2*(m1+m2-m2*(cos(phi1-phi2))**2))+(pphi1)/(l1**2*(m1+m2-m2*(cos(phi1-phi2))**2)),
         -1*(pphi1*cos(phi1-phi2))/(l1*l2*(m1+m2-m2*(cos(phi1-phi2))**2))+(pphi2*(m1+m2))/(l2**2*m2*(m1+m2-m2*(cos(phi1-phi2))**2)),
          -g*l1*(m1+m2)*sin(phi1) +  (pphi2**2*(m1+m2)*sin(phi1-phi2)*cos(phi1-phi2))/(l2**2*(m1+m2-m2*(cos(phi1-phi2))**2)**2)-
          (2*m2*pphi1*pphi2*sin(phi1-phi2)*(cos(phi1-phi2))**2)/(l1*l2*(m1+m2-m2*(cos(phi1-phi2))**2)**2)-
          (pphi1*pphi2*sin(phi1-phi2))/(l1*l2*(m1+m2-m2*(cos(phi1-phi2))**2))+
          (m2*pphi1**2*sin(phi1-phi2)*cos(phi1-phi2))/(l1**2*(m1+m2-m2*(cos(phi1-phi2))**2)**2), -g*l1*m2*sin(phi2)- 
          (pphi2**2*(m1+m2)*sin(phi1-phi2)*cos(phi1-phi2))/(l2**2*(m1+m2-m2*(cos(phi1-phi2))**2)**2)+
          (2*m2*pphi1*pphi2*sin(phi1-phi2)*(cos(phi1-phi2))**2)/(l1*l2*(m1+m2-m2*(cos(phi1-phi2))**2)**2)+
          (pphi1*pphi2*sin(phi1-phi2))/(l1*l2*(m1+m2-m2*(cos(phi1-phi2))**2))-
          (m2*pphi1**2*sin(phi1-phi2)*cos(phi1-phi2))/(l1**2*(m1+m2-m2*(cos(phi1-phi2))**2)**2)]

f0=[phi01,phi02,pphi01,pphi02] #condiciones iniciales
t=linspace(0,10,700)
sol=odeint(func,f0, t)
```

**Animación de las posiciones del péndulo doble**

```python
%matplotlib notebook

#Datos para graficar

x1=l1*sin(sol[:,0])
y1=-l1*cos(sol[:,0])

x2=x1+l2*sin(sol[:,1])
y2=y1-l2*cos(sol[:,1])

#Gráfica
fig=plt.figure()
ax=fig.gca()


#Función de actualización

def actualizar(i): 
    ax.clear()
    ax.plot([0,x1[i]],[0,y1[i]],'b-')
    ax.plot([x1[i],x2[i]],[y1[i],y2[i]],'b-')
    ax.plot(x1[i],y1[i],'ro')
    ax.plot(x2[i],y2[i],'ro')
    plt.title(str(round(t[i],3)))
    plt.xlim(-l1-l2,l1+l2) #para que no varíen los ejes
    plt.ylim(-l1-3*l2/2,0)
    plt.xlabel('$x [m]$')
    plt.ylabel('$y [m]$')
    plt.legend()
    plt.grid()
    
    
#Animación 

ani=animation.FuncAnimation(fig,actualizar,range(len(t)), interval=10, repeat=False)
#en donde está range se ponen los datos para meter en actualizar
#interval está en milisegundos, cada k milisegundos crea la gráfica
plt.show()
```

El número que aparece en el encabezado de la animación anterior es el tiempo en segundos

**Animación de la trayectoria en el espacio de fases para la  masa 1**

```python
%matplotlib notebook

#Datos para graficar


#Gráfica
fig1=plt.figure()
ax=fig1.gca()


#Función de actualización

def actualizar1(i): 
    ax.clear()
    ax.plot((sol[:,0])[:i],(sol[:,2])[:i],'b-')
    plt.title(str(round(t[i],3)))
    plt.xlim(min(sol[:,0]),max(sol[:,0])) #para que no varíen los ejes
    plt.ylim(min(sol[:,2]),max(sol[:,2]))
    plt.xlabel('$\phi_1$')
    plt.ylabel('$p_{\phi_1}$')
    plt.legend()
    plt.grid()
    
    
#Animación 

ani1=animation.FuncAnimation(fig1,actualizar1,range(len(t)), interval=1, repeat=False)
#en donde está range se ponen los datos para meter en actualizar
#interval está en milisegundos, cada k milisegundos crea la gráfica
plt.show()
```

Vemos en la gráfica anterior que bajo las condiciones inicialmente impustas, el movimiento de la masa $1$ no es una trayectoria cerrada en el espacio de las fases, eso quiere decir que no hay un periodo espacial definido para esta masa. Otra cosa que constrasta con las ideas del péndulo simple es que cuando el momento es cero  el valor del ángulo no siempre es máximo o mínimo esto ocurre porque la masa uno es "jalada" por la masa dos sin que haya llegado a el punto de máxima elongación. Sin embargo, una relación que se mantien con el péndulo simple es que cuando la masa pasa por el punto de equilibrio estable el momento (rapidéz) toma un valor aproximadamente extremal.

La trayectoria tampoco se cierra sobre sí misma, lo cual nos habla de una aperiodicidad espacial y temporal. 

**Animación de la evolución temporal de la energía del sistema**

```python
%matplotlib notebook

#Datos para graficar

pphi1=sol[:,2]
pphi2=sol[:,3]
phi1=sol[:,0]
phi2=sol[:,1]

E1=-1*((1/(2*m1))*l1**2*(pphi1)**2-m1*g*l1*cos(phi1))
H=-1*((pphi1**2)/(2*l1**2*((m1+m2)-m2*(cos(phi1-phi2))**2))-(pphi1*pphi2*cos(phi1-phi2))/(l1*l2*((m1+m2)-m2*(cos(phi1-phi2))**2))+(pphi2**2*(m1+m2))/(2*m2*l2**2*((m1+m2)-m2*(cos(phi1-phi2))**2))-(m1+m2)*g*l1*cos(phi1)-m2*g*l1*cos(phi2))
E2=H-E1
#Gráfica
fig3=plt.figure()
ax=fig3.gca()


#Función de actualización
  
def actualizar3(i): 
    ax.clear()
    ax.plot(t[:i],E1[:i],'b-', label='Energía m1')
    ax.plot(t[:i],E2[:i],'r-',label='Energía m2')
    ax.plot(t[:i],H[:i],'g-',label='Energía total')
    plt.title('Energías del sistema')
    plt.xlim(min(t),max(t)) #para que no varíen los ejes
    #plt.ylim(min(E),max(E))
    plt.xlabel(r'$t [s]$')
    plt.ylabel(r'$E [J] $')
    plt.legend()
    plt.grid()
    
    
#Animación 

ani3=animation.FuncAnimation(fig3,actualizar3,range(len(t)), interval=1, repeat=False)
#en donde está range se ponen los datos para meter en actualizar
#interval está en milisegundos, cada k milisegundos crea la gráfica
plt.show()
```

Como es de esperarse la energía del sistema se conserva así sea caótico, Las energías de ambas masas varían aumentando y dismonuyendo en el tiempo pero sumadas dan siempre la energía inicial, que es igual al hamiltoniano.
